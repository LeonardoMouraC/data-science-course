```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sinasc = pd.read_csv('SINASC_RO_2019.csv')
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
sns.barplot(x='CONSULTAS', y='APGAR5', data=sinasc)
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
plt.close('all')
plt.rc('figure', figsize=(15,15))
fig, axes = plt.subplots(2,2)

sns.countplot(ax= axes[0, 0], x='SEXO', data=sinasc)

sns.countplot(ax= axes[0, 1], x='GESTACAO', data=sinasc)

sns.countplot(ax= axes[1, 0], x='ESTCIVMAE', data=sinasc)

sns.countplot(ax= axes[1, 1], x='GESTACAO', data=sinasc)


axes[0,1].set_xticklabels(axes[0,1].get_xticklabels(),rotation = 30)
axes[1,1].set_xticklabels(axes[1,1].get_xticklabels(),rotation = 30)
```
